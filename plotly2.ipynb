```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(15)
data = {
    'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
    'Score': np.random.randint(50, 100, 200)
}
df = pd.DataFrame(data)
plt.figure(figsize=(8, 6))
sns.violinplot(x='Grade', y='Score', data=df)
plt.title('Distribution of Scores by Grade')
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)
sales_pivot = df.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='mean')
plt.figure(figsize=(10, 8))
sns.heatmap(sales_pivot, cmap='YlGnBu', annot=True, fmt=".0f")
plt.title('Sales Variation Across Days and Months')
plt.show()
```
