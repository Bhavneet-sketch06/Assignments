```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
np.random.seed(25)
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)
plt.figure(figsize=(8, 6))
plt.scatter(df['GDP'], df['Population'], s=df['Population'] * 10, alpha=0.5)
for i, country in enumerate(df['Country']):
    plt.text(df['GDP'][i], df['Population'][i], country, ha='center', va='center')

plt.xlabel('GDP')
plt.ylabel('Population')
plt.title('Country Population and GDP Bubble Chart')
plt.show()
```
